```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder, StandardScaler

import configparser
import utils
```

```python
config = configparser.ConfigParser()
config.read('../config.ini')
config = config['DEFAULT']
```

```python
training_data = pd.read_csv('../Data/train_potus_by_county.csv')
```

```python
# Let's take a look at the data
training_data.head()
```

```python
#Number of samples
len(training_data)
```

```python
# Missing Values
np.count_nonzero(training_data.isnull())
```

```python
# Extract features and labels
features_, labels_ = utils.getFeatureLabelColumns(training_data, config['ClassLabelsColumn'])
```

```python
# Distribution of class labels
winner_freq = labels_.value_counts()
winner_freq.plot(kind="bar", color="crimson", figsize=(8,6));
```

__NOTE__:  There is an uneven class distribution which may introduce significant bias in our model. Furthermore, the regular accuracy metric relies on even class distributions. Instead, we will rely on the f1 score to better judge the performance in model.ipynb.

```python
# Correlation between features/variables
corr = features_.corr()
```

```python
fig1, ax1 = plt.subplots(figsize=(14,10))
sns.heatmap(corr, cmap="BrBG", ax=ax1);
```

_Total Population_ and _Total Households_ are correlated as well as _Population Growth_ and _Household Growth_. There is a strong negative correlation between _Household Size_ and _Median Age_ as well as _% Owner Occupied Housing_ and _% Renter Housing_. Further feature engineering may be required.

```python
# Distribution of values for each feature
features_.hist(figsize=(15,12), bins=20);
```

__NOTE__: _Total Households_ and _Total Population_ do not have much variance. They may not provide much useful information.

```python
# Scale features to zero mean and unit variance. Encode labels into binary values.
features, scaler_ = utils.scaleUnitMean(features_)
labels, encoder_ = utils.labelEncode(labels_)
```

```python
# Compute and apply PCA transformation
pca_ = utils.computePCA(features, n_components=None)
pca_transformed_features = utils.applyPCA(features, pca_)
```

```python
# Plot the explained variance with each additional component
fig2, ax2 = plt.subplots(figsize=(8,5))
ax2.plot(1 - pca_.explained_variance_ratio_, c="#E91E63");
```

Roughly 95-98% of the variance in the data can be explained by 8 principle components. Let us plot the first two priciple components below.
