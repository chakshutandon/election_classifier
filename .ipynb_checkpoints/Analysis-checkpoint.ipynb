```python
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder, scale, StandardScaler
```

```python
training_data = pd.read_csv('train_potus_by_county.csv')
```

```python
# Let's take a look at the data
training_data.head()
```

```python
#Number of samples
len(training_data)
```

```python
# Missing Values
np.count_nonzero(training_data.isnull())
```

```python
# Extract features and labels
features = training_data.loc[:, training_data.columns != 'Winner']
labels = training_data['Winner']
```

```python
# Distribution of class labels
winner_freq = labels.value_counts()
winner_freq.plot(kind="bar", color="crimson", figsize=(8,6));
```

```python
# Correlation between features/variables
corr = features.corr()
```

```python
fig1, ax1 = plt.subplots(figsize=(14,10))
sns.heatmap(corr, cmap="BrBG", ax=ax1);
```

```python
# Distribution of values for each feature
features.hist(figsize=(15,12), bins=20);
```

```python
# Binary encode class labels and scale to unit variance (**TODO: Move Up)
le = LabelEncoder()
_training_data = training_data
_training_data['Winner'] = le.fit_transform(training_data['Winner'])

scaler = StandardScaler()
_training_data = scaler.fit_transform(_training_data)
# _training_data = scale(_training_data)
```

```python
pca = PCA()
_training_data = pca.fit_transform(_training_data);
```

```python
fig2, ax2 = plt.subplots(figsize=(8,5))
ax2.plot(pca.explained_variance_ratio_, c="#E91E63");
```

```python
fig3, ax3 = plt.subplots(figsize=(12,8))
pca1 = _training_data[:,0]
pca2 = _training_data[:,1]
color = list(map((lambda x: {"Barack Obama":"#CDDC39", "Mitt Romney":"#009688"}[x]), labels))
ax3.scatter(x=pca1, y=pca2, color=color, alpha=0.8);
```
